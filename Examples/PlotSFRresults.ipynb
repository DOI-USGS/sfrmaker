###Example workflow for plotting results from SFR text file output
(following a MODFLOW run)

```python
import sys
#sys.path.append('../Documents/GitHub/SFR')
sys.path.append('D:\ATLData\Documents\GitHub\SFR')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import flopy.utils.binaryfile as bf
from postproc import *

%matplotlib inline
```

###Create instance of Streamflow class  
Inputs are Mat1, Mat2, and SFR output text file

```python
path = ''
mat1 = path + 'SFRoutput/LPR_SFR_Mat1_20150123.csv'
mat2 = path + 'SFRoutput/LPR_SFR_Mat2_20150123.csv'
mfpath = path + 'LPR5_opt'

bedKmin = 1e-8 # K value to turn off conductance in collocated SFR reaches

sfrresults = Streamflow(Mat1=mat1, Mat2=mat2, streamflow_file=path  + 'LPR5_opt/SFR_output.dat')
```

####Read in the results

```python
df = sfrresults.read_streamflow_file()
```

####Create a dataframe from an individual segment and plot the streamflow
(streamflow is calculated as 0.5 * (Qin + Qout)

```python
seg34 = df.ix[df.segment == 34].copy()
seg34.index = seg34.reach
ax = seg34.Qstream.plot()
ax.set_ylabel('Streamflow')
```

####Plot the flow between the stream and aquifer

```python
ax = seg34.Qgw.plot()
ax.set_ylabel('Groundwater exchange')
```

####Plot the stage computed by SFR (icalc=1)

```python
seg34.stage.plot()
```

####Write a shapefile with the SFR results, to allow total streamflow; stream-aquifer interactions, and overland flow, etc. to be visualized spatially

```python
sfrresults.write_streamflow_shapefile(lines_shapefile='intermediate/new_streams_fragments.shp', node_col='node')
```

###The code below demonstrates how SFR results can be read from the CBB file using flopy; this is not implemented yet in SFRmaker

```python
cbbobj = bf.CellBudgetFile(mfpath + '/LPR_PP2.cbb')
cbbobj.list_records()
#rec = hdobj.get_data(kstpkper=(1, 50))
```

```python
sfr = cbbobj.get_data(text='  STREAM LEAKAGE')
np.shape(sfr[0][1])
```

###Get SFR exchange values at the row, column locations of Segment 34
first drop any collocated reaches that are zero-conductance so we can compare to cbb

```python
seg34_dropped = seg34.groupby(['row', 'column']).agg('sum').sort('reach')
rows = seg34_dropped.index[0][0]
columns = seg34_dropped.index[0][1]
seg34_dropped.index.get_level_values('row').values
```

```python
r = seg34_dropped.index.get_level_values('row').values -1
c = seg34_dropped.index.get_level_values('column').values -1

seg34_cbb = sfr[0][0][r, c]
```

```python
plt.plot(seg34_cbb)
```

```python
plt.plot(np.cumsum(seg34_cbb * -1), label='cbb file')
plt.hold(True)
plt.plot(np.cumsum(seg34_dropped.Qgw * -1), label='SFR .dat file')
plt.legend(loc=2)
plt.ylabel('Streamflow (cumulative flux)')
plt.xlabel('Reach')
plt.title('Segment 34')
plt.tight_layout()
plt.savefig('Segment_34_streamflow.pdf')
```

```python
plt.plot(seg34_cbb, label='cbb file')
plt.hold(True)
plt.plot(seg34_dropped.Qgw, label='SFR .dat file')
plt.legend(loc=3)
plt.ylabel('Groundwater flux to streamcells')
plt.xlabel('Reach')
plt.title('Segment 34')
plt.tight_layout()
plt.savefig('Segment_34_Qgw.pdf')
```
