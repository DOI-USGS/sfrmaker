###Intersect the polygon features(s) with the SFR cells
make a column in the polygon features dataframe that lists all intersecting SFR cells for each feature

```python
def wb_intersection(wb, sfr):
    intersects = [r.intersects(wb) for r in sfr.geometry]
    nodes = sfr.node[intersects].tolist()
    return nodes
```

###Renumber any sequence of reaches that broken into two or more segments
assumes that the reaches are sorted

```python
def renumber(seg, old_reaches):
    
    # correct any offset in first reach
    reach1 = np.min(old_reaches)
    old_reaches -= reach1 - 1

    new_reaches = old_reaches
    new_segments = np.array([seg])

    reach1 = 1
    gaps = np.diff(old_reaches)
    for i, diff in enumerate(gaps):
        if diff > 1:
            reach1 = old_reaches[i+1]
            seg +=1

        new_reaches[i+1] -= reach1 - 1    
        new_segments = np.append(new_segments, seg)
    
    return new_segments, new_reaches
```

###Make new segments and renumber their reaches

```python
import pandas as pd
m1 = pd.read_csv('/Users/aleaf/Documents/LPR/LPR_Calibration/2_from_ALEAF_SFR/SFRoutput/Mat1_elevs.csv')
m2 = pd.read_csv('/Users/aleaf/Documents/LPR/LPR_Calibration/2_from_ALEAF_SFR/SFRoutput/Mat2_elevs_edited.csv')

m1.sort(columns=['segment', 'reach'], inplace=True)
m2.sort('segment', inplace=True)
m2.index = m2.segment

# preserve old segments and reaches (for routing)
m1['old_segment'] = m1.segment
m1['old_reach'] = m1.reach

for seg in old_segments:
    
    # reaches that do not intersect the polygon feature
    inds1 = (~m1['node'].isin(new_seg_nodes)) & (m1.segment == seg)
    
    # reaches that intersect the polygon feature
    inds2 = m1['node'].isin(new_seg_nodes)
    
    # renumber the reaches
    for i, inds in enumerate([inds1, inds2]):
        
        firstnewseg = np.max(m1.segment) + 1
        old_reaches = m1.reach[inds].values

        # renumber these
        # the segments start at nseg, and then nseg + 1
        new_segments, new_reaches = renumber(firstnewseg + i, old_reaches)
        m1.loc[inds, 'segment'] = new_segments
        m1.loc[inds, 'reach'] = new_reaches
        
    # renumber the last new segment to the previous segment number
    # (can't have gaps in segment numbering; kind of clunky, but allows
    # the renumbering algorithm to be general
    lastnewseg = np.max(m1.segment)
    m1.loc[m1.segment == lastnewseg, 'segment'] = seg
    
    # update routing
    newsegs = np.unique(m1.ix[m1.old_segment == seg, 'segment'])
    old_outseg = int(m2.ix[seg, 'outseg'])
    for newseg in newsegs:
        
        # old reach number for reach 1 of the new segment
        old_reach_1 = m1.ix[(m1.segment == newseg) & (m1.reach == 1), 'old_reach'].values[0]
        
        if old_reach_1 > 1:
            # get the new segment number for next reach down from new segment reach 1
            outseg = m1.ix[(m1.old_segment == seg) & (m1.old_reach == old_reach_1 - 1), 'segment'].values[0]
        else:
            outseg = old_outseg # for segment begining at old reach 1, set outseg to previous outseg

        m2.loc[newseg, :] = m2.ix[seg, :] # copy all Mat2 info from old seg
        m2.loc[newseg, ['outseg', 'segment']] = outseg, newseg # update to new segment and outsegment numbers
```
