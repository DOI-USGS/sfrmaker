```python
def header(infile):
    ofp = open(infile)
    knt=0
    while True:
        try:
            line = ofp.readline().strip().split()
            int(line[0])
            break
        except:
            knt +=1
            continue
    return knt

h = header(sfr.streamflow_file)
```

```python
ofp = open(sfr.streamflow_file)
for i in np.arange(h):
    ofp.readline()
    
nreaches = len(sfr.m1)
sfrresults = {}
for i in np.arange(nreaches):
    line = ofp.readline().strip().split()
    l, r, c, s, reach = map(int, line[0:5])
    Qin, Qgw, Qout, Qovr, Qp, Qet, S, d, w, Cond, sb_slope = map(float, line[5:])
    
    Qstream = 0.5 * (Qin + Qout)
    
    if Qgw > 0:
        state = 'losing'
    elif Qgw < 0:
        state = 'gaining'
    elif Qstream == 0:
        state= 'dry'
    
    sfrresults[i] = {'layer': l,
                     'row': r,
                     'column': c,
                     'segment': s,
                     'reach': reach,
                     'Qin': Qin, 
                     'Qgw': Qgw,
                     'Qstream': Qstream,
                     'Qout': Qout, 
                     'Qovr': Qovr, 
                     'Qp': Qp, 
                     'Qet': Qet, 
                     'stage': S, 
                     'depth': d, 
                     'width': w, 
                     'cond': Cond, 
                     'sb_slope': sb_slope,
                     'state': state}
    
df = pd.DataFrame.from_dict(sfrresults, orient='index')
```
