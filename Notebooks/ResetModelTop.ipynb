####Use this same vector to reset the model tops to the streambed top
first make a vector of highest streambed values for each cell containing SFR (for collocated SFR cells)

```python
shared_cells = np.unique(sfr.m1.ix[sfr.m1.node.duplicated(), 'node'])
sfr.m1['highest_top'] = sfr.m1.top_streambed
for c in shared_cells:
    
    # select the collocated reaches for this cell
    df = sfr.m1[sfr.m1.node == c].sort('top_streambed', ascending=False)
    
    # set all of these reaches except the largest to not Dominant
    sfr.m1.loc[df.index, 'highest_top'] = np.max(df.top_streambed)
    
newtop = sfr.dis.top.array.copy()
newtop[sfr.m1.row.values-1, sfr.m1.column.values-1] = sfr.m1.top_streambed.values
```

####Inspect collocated SFR reaches that had different streambed elevations
(these have "highest_top" values that are different)

```python
sfr.m1.ix[sfr.m1.highest_top != sfr.m1.top_streambed, ['node', 'highest_top', 'top_streambed']].copy().sort('node')
```

####Now straighten out the other layers, removing any negative thicknesses

```python
# do layer 1 first
newbots = sfr.dis.botm.array.copy()
conflicts = newbots[0, :, :] > newtop - minimum_thickness
newbots[0, conflicts] = newtop[conflicts] - minimum_thickness

for i in range(sfr.dis.nlay - 1):
    conflicts = newbots[i+1, :, :] > (newbots[i, :, :] - 1)
    newbots[i+1, conflicts] = newbots[i, conflicts] - minimum_thickness
```

####Look at the reaches where the largest corrections were made

```python
sfr.m1['top_height'] = sfr.m1.model_top - sfr.m1.top_streambed
sfr.m1[sfr.m1.top_height > 0].sort(columns='top_height', ascending=False)
```
